## 3. GroupBy

Which `quality_level` are with higher `fixed_acidity` on average? Hint: use `groupby`

```python
# TODO
import pandas as pd
import numpy as np
quality_acidity = wine.groupby('quality_level').mean()['fixed_acidity']
print(quality_acidity)
higher_id = np.argmax(quality_acidity.values)
print("It is the '{}' quality wine has higher fixed acidity.".format(quality_acidity.index.tolist()[higher_id]))
```

Can you alo check for other features as what you did for `fixed_acidity`? Can you tell which one is more likely to reflect the `quality_level` of wine?


```python
# TODO
# TODO
# define function
from matplotlib import pyplot
%matplotlib inline

def check_quality(column):
    """ Check wine feature by quality level
    
    Arg:
       column (str): wine feature
    
    """
    
    # get mean by quality_level
    column_quality_mean = wine.groupby('quality_level').mean()[column]
    print(column_quality_mean)
    
    # plot distribution 
    pyplot.hist(wine[column][wine.quality_level=='low'], 
                alpha=0.2, label='quality low')
    pyplot.hist(wine[column][wine.quality_level=='high'], 
                alpha=0.5, label='quality high')
    pyplot.legend(loc='upper right')
    pyplot.title(column + ' vs quality_level')
    pyplot.xlabel(column)
    pyplot.ylabel('count')
    pyplot.show()
    print("="*60)

    
# check for each wine feature
for column in wine.drop(columns=['quality_level']).columns:
        check_quality(column)
        
# Based on the the distribution of each feature by quality_level, 
# there is no clear distinction between low and high quality wine except for the quality itself. 
```
